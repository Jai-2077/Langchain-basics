# 🧠 Memory in LangGraph

## Two Types of Memory
1. **Short-term (in-thread)**: conversation history within one session
2. **Long-term (cross-thread)**: facts persisted across all sessions

## Short-Term Memory
Managed by checkpointers — conversation history per thread_id.

## Long-Term Memory
Stored separately — user preferences, facts, summaries that persist across all conversations.


```python
# ── Long-Term Memory with LangGraph Store ────────────────────────────────
from langgraph.store.memory import InMemoryStore
from langgraph.checkpoint.memory import MemorySaver
from langgraph.graph import StateGraph, END
from langgraph.graph.message import add_messages
from langgraph.prebuilt import InjectedStore
from typing import TypedDict, Annotated
from langchain_core.messages import BaseMessage, HumanMessage
from langchain_openai import ChatOpenAI
import json

# Long-term store
store = InMemoryStore()

class AgentState(TypedDict):
    messages: Annotated[list[BaseMessage], add_messages]
    user_id: str

llm = ChatOpenAI(model='gpt-4o-mini')

def save_memory(state: AgentState, *, store: InMemoryStore) -> dict:
    """Extract and save facts from conversation to long-term memory."""
    if len(state['messages']) >= 2:
        last_human = next(
            (m for m in reversed(state['messages']) if m.type == 'human'), None
        )
        if last_human:
            # Save user message as a memory fact
            namespace = ('memories', state['user_id'])
            store.put(namespace, 'last_message', {'content': last_human.content})
    return {}

def recall_memory(state: AgentState, *, store: InMemoryStore) -> dict:
    """Retrieve memories and add context to conversation."""
    namespace = ('memories', state['user_id'])
    memories = store.search(namespace)
    print(f'Retrieved {len(memories)} memories for user {state["user_id"]}')
    return {}

print('Long-term memory pattern demonstrated.')
```
