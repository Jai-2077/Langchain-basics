# 🕸️ LangGraph Basics

## What is LangGraph?
LangGraph is a library for building **stateful, cyclic** AI workflows.

## Why LangGraph?
LCEL chains: A → B → C (linear, no loops)
LangGraph: Can loop back, branch, and maintain state

## Key Concepts
| Concept | Description |
|---------|-------------|
| Graph | The workflow structure |
| Node | A processing step (Python function) |
| Edge | Connection between nodes |
| State | Shared data passed between nodes |
| Conditional Edge | Branch based on state |
| Checkpointer | Saves state for resumption |

## Mental Model: State Machine
LangGraph is a **state machine** where:
- Each node can read and update state
- Edges determine what runs next
- Checkpointers save progress


```python
# ── Your First LangGraph ─────────────────────────────────────────────────
from typing import TypedDict, Annotated
from langgraph.graph import StateGraph, END
import operator

# ── Step 1: Define State ──
# State = shared data structure passed between nodes
class MyState(TypedDict):
    messages: list[str]       # conversation history
    step_count: int           # how many steps taken
    final_answer: str         # final output

# ── Step 2: Define Nodes ──
# Nodes are Python functions that take state and return updated state
def step_one(state: MyState) -> dict:
    print('Executing step_one...')
    return {
        'messages': state['messages'] + ['Step 1 complete'],
        'step_count': state['step_count'] + 1,
    }

def step_two(state: MyState) -> dict:
    print('Executing step_two...')
    return {
        'messages': state['messages'] + ['Step 2 complete'],
        'step_count': state['step_count'] + 1,
        'final_answer': f'Done after {state["step_count"] + 1} steps!'
    }

# ── Step 3: Build Graph ──
graph = StateGraph(MyState)

# Add nodes
graph.add_node('step_one', step_one)
graph.add_node('step_two', step_two)

# Add edges
graph.set_entry_point('step_one')  # START → step_one
graph.add_edge('step_one', 'step_two')  # step_one → step_two
graph.add_edge('step_two', END)  # step_two → END

# Compile
app = graph.compile()

# ── Step 4: Run ──
initial_state = MyState(messages=[], step_count=0, final_answer='')
result = app.invoke(initial_state)

print('\nFinal state:')
print(f'  Messages: {result["messages"]}')
print(f'  Steps: {result["step_count"]}')
print(f'  Answer: {result["final_answer"]}')
```

```python
# ── Visualizing the Graph ─────────────────────────────────────────────
# LangGraph can generate ASCII and visual diagrams of your graph.

try:
    # ASCII diagram
    print(app.get_graph().draw_ascii())
except:
    print('Graph structure:')
    print('  START → step_one → step_two → END')

# Get graph nodes and edges programmatically
graph_obj = app.get_graph()
print('\nNodes:', list(graph_obj.nodes.keys()))
print('Edges:', [(e.source, e.target) for e in graph_obj.edges])
```
