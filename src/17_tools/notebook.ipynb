# 🛠️ Tools

## What are Tools?
**Tools** are Python functions that an LLM can **decide to call**.

LLMs know about the world up to their training cutoff.
Tools let LLMs:
- Search the internet
- Run Python code
- Query databases
- Call APIs
- Read/write files

## Mental Model: Human with Assistants
An executive (LLM) has assistants (tools):
- "Search the web for this" → WebSearch tool
- "Calculate this formula" → Calculator tool
- "Send this email" → Email tool

The executive DECIDES which assistant to call and WHEN.

## Tool Anatomy
```python
@tool
def search_web(query: str) -> str:
    """Search the internet for current information."""
    # implementation
    return results
```

LangChain reads the:
- **Name** → `search_web` (how LLM refers to it)
- **Docstring** → LLM uses this to understand WHEN to call it
- **Parameters** → what the LLM needs to provide
- **Return type** → what the LLM receives back


````python
# ── Creating Tools with @tool Decorator ─────────────────────────────────
from langchain_core.tools import tool
from typing import Annotated

@tool
def add_numbers(a: float, b: float) -> float:
    """Add two numbers together. Use when you need to perform addition."""
    return a + b

@tool
def multiply_numbers(a: float, b: float) -> float:
    """Multiply two numbers. Use when you need to compute a product."""
    return a * b

@tool
def get_weather(city: str) -> str:
    """Get current weather for a city. Returns temperature and conditions."""
    # Simulated — in production, call a weather API
    weather_data = {
        'London': 'Rainy, 12°C',
        'Paris': 'Sunny, 22°C',
        'Tokyo': 'Cloudy, 18°C',
    }
    return weather_data.get(city, f'Weather data not available for {city}')

# Inspect tool metadata
print('Tool name:', add_numbers.name)
print('Tool description:', add_numbers.description)
print('Tool schema:', add_numbers.args_schema.model_json_schema())

# Call tool directly
result = add_numbers.invoke({'a': 5, 'b': 3})
print('\nDirect call result:', result)
````

````python
# ── Binding Tools to an LLM ─────────────────────────────────────────────
# WHY: The LLM needs to KNOW about available tools.
# bind_tools() sends tool schemas to the LLM so it knows when/how to call them.

from langchain_openai import ChatOpenAI
from langchain_core.tools import tool

@tool
def calculate(expression: str) -> str:
    """Evaluate a mathematical expression. Input should be a valid Python math expression."""
    try:
        result = eval(expression, {'__builtins__': {}}, {})
        return str(result)
    except Exception as e:
        return f'Error: {e}'

@tool
def get_current_time() -> str:
    """Get the current date and time."""
    from datetime import datetime
    return datetime.now().strftime('%Y-%m-%d %H:%M:%S')

tools = [calculate, get_current_time]

llm = ChatOpenAI(model='gpt-4o-mini')
llm_with_tools = llm.bind_tools(tools)

# Ask something that requires a tool
response = llm_with_tools.invoke('What is 42 * 17?')
print('Response type:', type(response).__name__)
print('Tool calls requested:', response.tool_calls)
print('Content:', response.content)
````
