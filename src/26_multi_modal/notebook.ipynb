# 🖼️ Multi-Modal LLMs

## What is Multi-Modal?
Multi-modal LLMs can process **multiple types of input**:
- Text
- Images
- Audio (some models)
- Video (some models)

## Supported Models
- GPT-4o: text + images
- Claude 3.x: text + images
- Gemini: text + images + audio + video

## Use Cases
- Image analysis
- Document OCR
- Chart/graph reading
- Visual QA
- Medical imaging analysis


```python
# ── Image Analysis with GPT-4o ─────────────────────────────────────────
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
import base64

# Method 1: URL-based (no download needed)
llm = ChatOpenAI(model='gpt-4o', max_tokens=500)

# Build message with image URL
message = HumanMessage(content=[
    {'type': 'text', 'text': 'Describe what you see in this image in detail.'},
    {
        'type': 'image_url',
        'image_url': {
            'url': 'https://upload.wikimedia.org/wikipedia/commons/thumb/4/47/PNG_transparency_demonstration_1.png/280px-PNG_transparency_demonstration_1.png',
            'detail': 'low'  # 'low', 'high', or 'auto'
        }
    }
])

response = llm.invoke([message])
print('Image Analysis:', response.content)

# Method 2: Base64 encoded (for local files)
def analyze_local_image(image_path: str, question: str) -> str:
    with open(image_path, 'rb') as f:
        image_data = base64.b64encode(f.read()).decode()

    message = HumanMessage(content=[
        {'type': 'text', 'text': question},
        {'type': 'image_url', 'image_url': {'url': f'data:image/jpeg;base64,{image_data}'}}
    ])
    return llm.invoke([message]).content

print('\nBase64 function ready. Call analyze_local_image(path, question)')
```
