# ⚡ Streaming

## What is Streaming?
Instead of waiting for the full LLM response, **stream** tokens as they are generated.

Without streaming: User waits 5-10 seconds. Then sees full response.
With streaming: User sees text appearing word by word. Much better UX!

## Why Streaming Matters
- ChatGPT-like user experience
- Perceived latency is much lower
- Can cancel early if response is wrong

## LCEL + Streaming
All LCEL chains support streaming for free via `.stream()`.


```python
# ── Basic Streaming ──────────────────────────────────────────────────────
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

chain = (
    ChatPromptTemplate.from_template('Write a 3 sentence story about {topic}')
    | ChatOpenAI(model='gpt-4o-mini', temperature=0.7)
    | StrOutputParser()
)

print('Streaming response (tokens appear one by one):')
print('─' * 50)

full_text = ''
for chunk in chain.stream({'topic': 'a robot learning to paint'}):
    print(chunk, end='', flush=True)  # flush=True forces immediate display
    full_text += chunk

print('\n' + '─' * 50)
print(f'Total chars streamed: {len(full_text)}')
```

```python
# ── Async Streaming ────────────────────────────────────────────────────
# WHY: In web apps (FastAPI), async streaming is essential.
# It allows serving multiple users without blocking.

import asyncio
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

chain = (
    ChatPromptTemplate.from_template('Explain {topic} in 2 sentences')
    | ChatOpenAI(model='gpt-4o-mini')
    | StrOutputParser()
)

async def stream_response(topic: str):
    print(f'Streaming about: {topic}')
    async for chunk in chain.astream({'topic': topic}):
        print(chunk, end='', flush=True)
    print()

asyncio.run(stream_response('quantum computing'))
```
