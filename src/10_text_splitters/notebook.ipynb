# ✂️ Text Splitters

## Why Split Text?
LLMs have **context window limits** (e.g., 128K tokens).
A 500-page book exceeds any context window.
Solution: **chunk** the document into smaller pieces.

## The Goldilocks Problem
- Too large → exceeds context window, loses important info
- Too small → loses context, fragmented meaning
- Just right → preserves meaning, fits context window

## Chunking Strategy
```
Original Document (10,000 words)
        ↓
Text Splitter
        ↓
[Chunk 1 (200 words), Chunk 2 (200 words), ..., Chunk 50 (200 words)]
        ↓
Embed each chunk
        ↓
Store in vector database
```

## Key Parameters
- **chunk_size**: max chars/tokens per chunk
- **chunk_overlap**: how many chars overlap between adjacent chunks
  - WHY overlap? Prevents important context from being split across boundaries

## Splitter Types
| Splitter | Split Strategy |
|----------|---------------|
| CharacterTextSplitter | By character count |
| RecursiveCharacterTextSplitter | Smart recursive (RECOMMENDED) |
| TokenTextSplitter | By tokens |
| MarkdownTextSplitter | By markdown headers |
| SemanticChunker | By semantic meaning |


````python
# ── RecursiveCharacterTextSplitter (Recommended) ────────────────────────
# WHY 'Recursive'? It tries splitting by paragraphs first,
# then sentences, then words, then characters.
# This preserves meaning better than simple character splitting.

from langchain_text_splitters import RecursiveCharacterTextSplitter

text = '''\nLangChain is a framework for developing applications powered by language models.\n\nIt provides tools and abstractions to improve the customization, accuracy, and relevancy \nof the information the models generate. \n\nLangChain offers Python and JavaScript libraries. The Python package is the most mature.\n\nKey concepts include: chains, agents, tools, memory, and callbacks.\n'''

splitter = RecursiveCharacterTextSplitter(
    chunk_size=150,       # max 150 chars per chunk
    chunk_overlap=30,     # 30 char overlap between chunks
    length_function=len,  # use len() to measure size
    separators=['\n\n', '\n', '. ', ' ', ''],  # try these in order
)

chunks = splitter.split_text(text)
print(f'Original length: {len(text)} chars')
print(f'Number of chunks: {len(chunks)}')
print(f'\nChunk sizes: {[len(c) for c in chunks]}')
for i, chunk in enumerate(chunks):
    print(f'\n── Chunk {i+1} ──\n{chunk}')
````

````python
# ── Splitting Documents (not just strings) ─────────────────────────────
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document

# Create sample documents
docs = [
    Document(
        page_content='Python is a high-level, general-purpose programming language. '
                     'Its design philosophy emphasizes code readability. '
                     'Python is dynamically typed and garbage-collected. '
                     'It supports multiple programming paradigms.',
        metadata={'source': 'python_wiki.txt', 'page': 1}
    )
]

splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20)
split_docs = splitter.split_documents(docs)

print(f'Input: {len(docs)} document(s)')
print(f'Output: {len(split_docs)} chunk(s)')
print('\nImportant: metadata is PRESERVED in each chunk!')
for i, doc in enumerate(split_docs):
    print(f'\nChunk {i+1}:')
    print(f'  Content: {doc.page_content[:80]}...')
    print(f'  Metadata: {doc.metadata}')  # source is preserved!
````

````python
# ── Overlap Visualization ─────────────────────────────────────────────
# WHY overlap? Without it, important sentences get cut in half.

text = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

from langchain_text_splitters import CharacterTextSplitter

no_overlap = CharacterTextSplitter(chunk_size=10, chunk_overlap=0, separator='')
with_overlap = CharacterTextSplitter(chunk_size=10, chunk_overlap=3, separator='')

print('WITHOUT overlap:')
for chunk in no_overlap.split_text(text):
    print(f'  {chunk}')

print('\nWITH overlap (3 chars):')
for chunk in with_overlap.split_text(text):
    print(f'  {chunk}')
print('\nNotice each chunk shares 3 chars with the previous chunk.')
print('In real text, this prevents sentences from being cut across boundaries.')
````
