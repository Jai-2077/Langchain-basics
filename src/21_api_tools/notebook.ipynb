# 🌐 API Tools

## What are API Tools?
API tools let LLMs interact with external services:
- Weather APIs
- News APIs
- Stock prices
- Search engines
- REST APIs

## Building Custom API Tools
Any API can become a LangChain tool.


```python
# ── Custom API Tool Pattern ─────────────────────────────────────────────
import httpx
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import create_react_agent

@tool
def get_joke(category: str = 'general') -> str:
    """Fetch a random joke from the internet. Categories: general, programming, dark."""
    try:
        # Using JokeAPI (free, no auth required)
        url = f'https://v2.jokeapi.dev/joke/{category}?type=single'
        response = httpx.get(url, timeout=5)
        data = response.json()
        if data.get('error'):
            return 'Could not fetch joke.'
        return data.get('joke', 'No joke found')
    except Exception as e:
        return f'Error fetching joke: {e}'

@tool
def get_random_fact() -> str:
    """Get a random interesting fact."""
    try:
        response = httpx.get('https://uselessfacts.jsph.pl/api/v2/facts/random', timeout=5)
        return response.json().get('text', 'No fact available')
    except:
        return 'Could not fetch fact'

tools = [get_joke, get_random_fact]
agent = create_react_agent(ChatOpenAI(model='gpt-4o-mini'), tools)

result = agent.invoke({'messages': [('user', 'Tell me a programming joke')]})
print(result['messages'][-1].content)
```
