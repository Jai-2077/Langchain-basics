# 🚀 Deployment

## Production Checklist
- [ ] Environment variables secured
- [ ] Rate limiting implemented
- [ ] Error handling and retries
- [ ] Monitoring (LangSmith)
- [ ] Streaming enabled
- [ ] Caching (reduce API costs)
- [ ] Authentication
- [ ] Load testing done

## Deployment Options
| Option | Best For |
|--------|---------|
| FastAPI | Custom APIs |
| LangServe | Quick LangChain deployment |
| Docker | Containerized deployment |
| AWS Lambda | Serverless |
| Cloud Run | Auto-scaling |


```python
# ── Production FastAPI App ─────────────────────────────────────────────
production_app = '''
# production_app.py
import os
import logging
from contextlib import asynccontextmanager
from fastapi import FastAPI, HTTPException, Request
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv

load_dotenv()
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Global chain (initialized once at startup)
chain = None

@asynccontextmanager
async def lifespan(app: FastAPI):
    global chain
    logger.info("Initializing LangChain...")
    chain = (
        ChatPromptTemplate.from_template("{question}")
        | ChatOpenAI(model="gpt-4o-mini").with_retry(stop_after_attempt=3)
        | StrOutputParser()
    )
    logger.info("Ready!")
    yield
    logger.info("Shutting down...")

app = FastAPI(title="LangChain API", lifespan=lifespan)
app.add_middleware(CORSMiddleware, allow_origins=["*"], allow_methods=["*"])

class QueryRequest(BaseModel):
    question: str

@app.post("/query")
async def query(request: QueryRequest):
    try:
        result = await chain.ainvoke({"question": request.question})
        return {"answer": result, "status": "success"}
    except Exception as e:
        logger.error(f"Error: {e}")
        raise HTTPException(status_code=500, detail=str(e))

@app.get("/health")
async def health(): return {"status": "healthy"}
'''
print(production_app)
```
