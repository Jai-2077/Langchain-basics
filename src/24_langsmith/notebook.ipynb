# 🔭 LangSmith

## What is LangSmith?
**LangSmith** is Anthropic's observability platform for LangChain applications.

## What it Does
- **Traces** every LLM call, tool call, chain step
- Shows exact prompts sent and responses received
- Tracks latency and token usage
- Enables prompt testing and evaluation
- Production monitoring

## Why Observability?
Without observability, debugging AI apps is like:
- Flying blind
- No logs, no metrics
- Can't tell why the LLM gave a bad answer

With LangSmith:
- See exactly what prompt was sent
- See exactly what the LLM returned
- Track which prompts work best
- Monitor production quality


```python
# ── Setting Up LangSmith ─────────────────────────────────────────────────
import os
from dotenv import load_dotenv
load_dotenv()

# Required environment variables
# Add these to your .env file:
# LANGCHAIN_TRACING_V2=true
# LANGCHAIN_API_KEY=your_key_here
# LANGCHAIN_PROJECT=my-project

os.environ.setdefault('LANGCHAIN_TRACING_V2', 'true')
os.environ.setdefault('LANGCHAIN_PROJECT', 'langchain-mastercourse')

print('LangSmith configuration:')
print(f'  Tracing: {os.getenv("LANGCHAIN_TRACING_V2", "not set")}')
print(f'  Project: {os.getenv("LANGCHAIN_PROJECT", "not set")}')
print(f'  API Key: {"set" if os.getenv("LANGCHAIN_API_KEY") else "NOT SET"}')

# Once configured, ALL chain calls are automatically traced!
# No code changes needed.

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

chain = (
    ChatPromptTemplate.from_template('Summarize {topic} in 1 sentence.')
    | ChatOpenAI(model='gpt-4o-mini')
    | StrOutputParser()
)

# This call will appear in LangSmith dashboard
result = chain.invoke({'topic': 'machine learning'})
print('Result:', result)
print('\nCheck LangSmith dashboard: https://smith.langchain.com')
```
