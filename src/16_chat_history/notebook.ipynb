# 💾 Chat History

## What is Chat History?
**Chat history** = stored record of past messages in a conversation.

## Storage Options
| Storage | Package | Best For |
|---------|---------|---------|
| In-Memory | langchain-core | Development/testing |
| Redis | langchain-community | Production, fast |
| PostgreSQL | langchain-community | Production, relational |
| MongoDB | langchain-mongodb | Production, document |
| DynamoDB | langchain-aws | AWS deployments |
| File System | langchain-community | Simple persistence |

## Why Persist Chat History?
- Users return to continue conversations
- Audit trail for compliance
- Analytics on user interactions
- Debug production issues


```python
# ── InMemoryChatMessageHistory ─────────────────────────────────────────
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.messages import HumanMessage, AIMessage

history = InMemoryChatMessageHistory()

# Add messages
history.add_user_message('Hello!')
history.add_ai_message('Hi there! How can I help?')
history.add_user_message('What is LangChain?')

# Inspect
print('Messages in history:')
for msg in history.messages:
    print(f'  [{msg.type}]: {msg.content}')

# Clear
history.clear()
print(f'\nAfter clear: {len(history.messages)} messages')
```

```python
# ── File-Based Chat History ─────────────────────────────────────────────
# WHY: Survives program restarts. Useful for development.

from langchain_community.chat_message_histories import FileChatMessageHistory

history = FileChatMessageHistory('/tmp/chat_history.json')
history.clear()  # Start fresh

history.add_user_message('Remember: I prefer Python over JavaScript.')
history.add_ai_message('Got it! I will use Python in my examples.')

print('Saved to file!')

# Reload from file (simulates program restart)
reloaded = FileChatMessageHistory('/tmp/chat_history.json')
print('Reloaded messages:')
for msg in reloaded.messages:
    print(f'  [{msg.type}]: {msg.content}')
```
