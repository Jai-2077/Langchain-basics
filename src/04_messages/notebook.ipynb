# 💬 Messages — The Language of LLMs

## What Are Messages?
Modern LLMs (GPT-4, Claude, Gemini) don't take a single string as input.
They take a **list of structured messages**, each with a role.

## Why Roles?
The LLM needs to know WHO said what:
- **System** → Instructions, persona, rules (from the developer)
- **Human** → User's input
- **AI** → Previous LLM responses (for multi-turn conversations)
- **Tool** → Results from tool/function calls

## Mental Model: Email Thread
```
SystemMessage  = Company email policy document
HumanMessage   = Email you wrote
AIMessage      = Reply you received
HumanMessage   = Your follow-up
AIMessage      = Their follow-up
```

## Why Not Just Strings?
Before chat models, LLMs took raw strings. Problems:
- No way to separate system instructions from user input
- No multi-turn conversation support
- LLM couldn't distinguish who said what


````python
# ── Message Types ────────────────────────────────────────────────────────
from langchain_core.messages import (
    SystemMessage,    # Developer instructions — shapes LLM behavior
    HumanMessage,     # User input
    AIMessage,        # LLM response (used in multi-turn chat)
    ToolMessage,      # Result from tool/function call
    BaseMessage,      # Parent class of all messages
)

# Create each type
system = SystemMessage(content='You are a helpful Python tutor.')
human = HumanMessage(content='What is a list comprehension?')
ai = AIMessage(content='A list comprehension is a concise way to create lists...')

# Inspect internal structure
print('── SystemMessage ──')
print(f'  type: {type(system).__name__}')
print(f'  content: {system.content}')
print(f'  role: {system.type}')  # 'system'

print('\n── HumanMessage ──')
print(f'  type: {type(human).__name__}')
print(f'  role: {human.type}')  # 'human'

print('\n── AIMessage ──')
print(f'  type: {type(ai).__name__}')
print(f'  role: {ai.type}')  # 'ai'
print(f'  tool_calls: {ai.tool_calls}')  # [] when no tools used
````

````python
# ── Messages in Action: Multi-Turn Conversation ─────────────────────────
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)

# Build conversation history manually
messages = [
    SystemMessage(content='You are a concise Python tutor. Keep answers under 2 sentences.'),
    HumanMessage(content='What is a decorator?'),
]

# Turn 1
response1 = llm.invoke(messages)
print('Turn 1 Response:', response1.content)

# Add AI response to history + new human message
messages.append(response1)  # AIMessage goes back into history
messages.append(HumanMessage(content='Show me a simple example.'))

# Turn 2 — LLM now has full context
response2 = llm.invoke(messages)
print('\nTurn 2 Response:', response2.content)

print(f'\nTotal messages in conversation: {len(messages) + 1}')
print('This is how ChatGPT "remembers" previous messages!')
````

````python
# ── Serialization: Messages as Dicts ───────────────────────────────────
# WHY: Messages must be serialized for storage, APIs, and databases.

from langchain_core.messages import HumanMessage, messages_to_dict, messages_from_dict

msgs = [
    HumanMessage(content='Hello!'),
]

# Convert to dict (for storage)
as_dicts = messages_to_dict(msgs)
print('As dicts (for DB storage):')
print(as_dicts)

# Convert back from dict
restored = messages_from_dict(as_dicts)
print('\nRestored messages:')
print(restored)
print('Types preserved:', [type(m).__name__ for m in restored])
````
