# 📊 Structured Output

## The Problem
LLMs return text. Applications need structured data.

## Solutions (in order of reliability)
1. **with_structured_output()** — uses tool calling, most reliable
2. **PydanticOutputParser** — uses format instructions, less reliable
3. **JsonOutputParser** — basic JSON parsing
4. **Manual parsing** — fragile, not recommended

## when to Use
- Extracting structured data from text
- Form filling
- Classification
- Entity extraction


```python
# ── with_structured_output: The Modern Way ───────────────────────────────
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field
from typing import Optional

class JobPosting(BaseModel):
    title: str = Field(description='Job title')
    company: str = Field(description='Company name')
    location: str = Field(description='Job location (city or remote)')
    salary_min: Optional[int] = Field(default=None, description='Min salary USD')
    salary_max: Optional[int] = Field(default=None, description='Max salary USD')
    required_skills: list[str] = Field(description='Required technical skills')
    remote: bool = Field(description='Is the position remote?')

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)
structured_llm = llm.with_structured_output(JobPosting)

job_text = '''
We are hiring a Senior Python Engineer at TechCorp in San Francisco (remote OK).
Compensation: $140,000 - $180,000. Must know Python, FastAPI, PostgreSQL, Docker.
'''

result = structured_llm.invoke(job_text)
print('Extracted JobPosting:')
print(f'  Title: {result.title}')
print(f'  Company: {result.company}')
print(f'  Salary: ${result.salary_min:,} - ${result.salary_max:,}')
print(f'  Remote: {result.remote}')
print(f'  Skills: {result.required_skills}')
```
