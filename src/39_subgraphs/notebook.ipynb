# 🗂️ Subgraphs

## What are Subgraphs?
A **subgraph** is a complete LangGraph graph used as a **node** inside another graph.

## Why Subgraphs?
- **Modularity**: break complex graphs into manageable pieces
- **Reuse**: same subgraph in multiple parent graphs
- **Encapsulation**: hide implementation details
- **Team collaboration**: different teams own different subgraphs

## Mental Model: Functions in Code
```python
# Regular function (reusable, encapsulated)
def process_document(doc): ...

# Subgraph (reusable, encapsulated)
document_graph = build_document_graph()
parent_graph.add_node('process_doc', document_graph)
```


````python
# ── Building and Using a Subgraph ────────────────────────────────────────
from typing import TypedDict, Annotated
from langgraph.graph import StateGraph, END

# ── Subgraph: handles document processing ──
class DocState(TypedDict):
    text: str
    word_count: int
    cleaned: bool

def clean_text(state: DocState) -> dict:
    cleaned = state['text'].strip().lower()
    return {'text': cleaned, 'cleaned': True}

def count_words(state: DocState) -> dict:
    return {'word_count': len(state['text'].split())}

doc_graph = StateGraph(DocState)
doc_graph.add_node('clean', clean_text)
doc_graph.add_node('count', count_words)
doc_graph.set_entry_point('clean')
doc_graph.add_edge('clean', 'count')
doc_graph.add_edge('count', END)
doc_subgraph = doc_graph.compile()  # compile the subgraph

# ── Parent graph: uses subgraph as a node ──
class PipelineState(TypedDict):
    text: str
    word_count: int
    cleaned: bool
    pipeline_done: bool

def finalize(state: PipelineState) -> dict:
    return {'pipeline_done': True}

parent = StateGraph(PipelineState)
parent.add_node('doc_processor', doc_subgraph)  # subgraph as node!
parent.add_node('finalize', finalize)
parent.set_entry_point('doc_processor')
parent.add_edge('doc_processor', 'finalize')
parent.add_edge('finalize', END)
pipeline = parent.compile()

result = pipeline.invoke({'text': '  Hello World from LangGraph!  ', 'word_count': 0, 'cleaned': False, 'pipeline_done': False})
print('Result:', result)
````
