# 🗃️ SQL Agents

## What is a SQL Agent?
A SQL agent converts **natural language questions** into **SQL queries** and returns results.

```
User: "How many users signed up last month?"
Agent: SELECT COUNT(*) FROM users WHERE created_at >= '2024-01-01'
Result: 1,247 users
```

## Use Cases
- Business intelligence without SQL knowledge
- Automated reporting
- Data exploration chatbots
- Executive dashboards


````python
# ── SQL Database Agent ───────────────────────────────────────────────────
from langchain_community.utilities import SQLDatabase
from langchain_openai import ChatOpenAI
from langchain_community.agent_toolkits import create_sql_agent
import sqlite3
import os

# Create sample SQLite database
db_path = '/tmp/company.db'
conn = sqlite3.connect(db_path)
conn.executescript('''
    CREATE TABLE IF NOT EXISTS employees (
        id INTEGER PRIMARY KEY,
        name TEXT,
        department TEXT,
        salary REAL,
        hire_date TEXT
    );
    DELETE FROM employees;
    INSERT INTO employees VALUES
        (1, 'Alice', 'Engineering', 95000, '2022-01-15'),
        (2, 'Bob', 'Marketing', 75000, '2021-06-01'),
        (3, 'Carol', 'Engineering', 105000, '2020-03-10'),
        (4, 'Dave', 'HR', 65000, '2023-09-01'),
        (5, 'Eve', 'Engineering', 115000, '2019-11-20');
''')
conn.commit()
conn.close()

# Connect LangChain to the database
db = SQLDatabase.from_uri(f'sqlite:///{db_path}')
print('Tables:', db.get_usable_table_names())

# Create SQL agent
llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)
agent = create_sql_agent(llm, db=db, verbose=True)

# Ask natural language questions
result = agent.invoke({'input': 'What is the average salary in Engineering?'})
print('\nAnswer:', result['output'])
````
