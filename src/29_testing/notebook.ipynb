# 🧪 Testing LangChain Applications

## Why Testing is Hard
LLMs are non-deterministic. Testing them requires:
- Mocking LLM responses
- Evaluating output quality (not just equality)
- Testing at multiple levels

## Testing Levels
1. **Unit tests** — test individual components with mocked LLMs
2. **Integration tests** — test chains end-to-end
3. **Evaluation** — assess output quality with LLM-as-judge


```python
# ── Mocking LLMs for Unit Tests ─────────────────────────────────────────
from langchain_core.language_models.fake import FakeListChatModel
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
import unittest

# FakeListChatModel returns predefined responses (no API calls!)
fake_llm = FakeListChatModel(
    responses=['Paris', 'Tokyo', 'London']  # Cycles through these
)

chain = ChatPromptTemplate.from_template('{q}') | fake_llm | StrOutputParser()

class TestCapitalChain(unittest.TestCase):
    def test_returns_string(self):
        result = chain.invoke({'q': 'Capital of France?'})
        self.assertIsInstance(result, str)
        self.assertEqual(result, 'Paris')

    def test_multiple_calls(self):
        r1 = chain.invoke({'q': 'q1'})
        r2 = chain.invoke({'q': 'q2'})
        self.assertEqual(r1, 'Tokyo')
        self.assertEqual(r2, 'London')

# Run tests
suite = unittest.TestLoader().loadTestsFromTestCase(TestCapitalChain)
runner = unittest.TextTestRunner(verbosity=2)
result = runner.run(suite)
```
