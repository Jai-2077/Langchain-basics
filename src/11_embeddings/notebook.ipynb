# 🔢 Embeddings

## What are Embeddings?
An **embedding** converts text into a list of numbers (a vector).

```
"Hello, world!" → [0.23, -0.45, 0.67, ..., 0.12]  (1536 numbers)
```

## Why Vectors?
Computers can't understand words. But they can:
- Measure **distance** between vectors
- Find **similar** vectors
- Perform **arithmetic** on vectors

## The Magic: Semantic Similarity
```
"dog" → [0.2, 0.8, ...]
"puppy" → [0.19, 0.81, ...]  ← similar!
"quantum" → [0.9, 0.1, ...]  ← very different
```

Similar MEANING = similar vectors.

## Mental Model: GPS Coordinates
- Paris: lat 48.8, lon 2.3
- Lyon: lat 45.7, lon 4.8  ← nearby France city
- Tokyo: lat 35.6, lon 139.6 ← far away

Embeddings are "meaning coordinates" in 1536-dimensional space.

## Why Embeddings Enable RAG
1. Embed all your documents → store in vector DB
2. User asks a question → embed the question
3. Find documents with SIMILAR vectors to the question
4. Pass those documents to LLM as context


````python
# ── Creating Embeddings ─────────────────────────────────────────────────
from langchain_openai import OpenAIEmbeddings

# OpenAI text-embedding-3-small: fast, cheap, 1536 dimensions
embeddings = OpenAIEmbeddings(model='text-embedding-3-small')

# Embed a single text
vector = embeddings.embed_query('Hello, LangChain!')
print(f'Type: {type(vector).__name__}')  # list
print(f'Dimensions: {len(vector)}')      # 1536
print(f'First 5 values: {vector[:5]}')
print(f'Range: [{min(vector):.3f}, {max(vector):.3f}]')
````

````python
# ── Semantic Similarity with Cosine Distance ────────────────────────────
# WHY cosine similarity? It measures ANGLE between vectors,
# not distance. This works better for high-dimensional text vectors.
# Cosine similarity: 1.0 = identical, 0 = unrelated, -1 = opposite.

import numpy as np
from langchain_openai import OpenAIEmbeddings

def cosine_similarity(v1: list, v2: list) -> float:
    v1, v2 = np.array(v1), np.array(v2)
    return float(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))

embeddings = OpenAIEmbeddings(model='text-embedding-3-small')

sentences = [
    'I love programming in Python',
    'Python is my favorite coding language',  # similar!
    'The sky is blue and full of clouds',     # different
]

vecs = embeddings.embed_documents(sentences)
ref = vecs[0]  # 'I love programming in Python'

print('Similarity to "I love programming in Python":')
for i, (sent, vec) in enumerate(zip(sentences, vecs)):
    sim = cosine_similarity(ref, vec)
    bar = '█' * int(sim * 20)
    print(f'  [{sim:.4f}] {bar} {sent[:50]}')
````

````python
# ── Embedding Documents for RAG ─────────────────────────────────────────
from langchain_openai import OpenAIEmbeddings
from langchain_core.documents import Document

embeddings = OpenAIEmbeddings(model='text-embedding-3-small')

docs = [
    Document(page_content='LangChain helps build LLM applications.'),
    Document(page_content='Embeddings convert text to vectors.'),
    Document(page_content='Vector databases store embeddings.'),
]

# Embed all documents at once (batched — more efficient)
texts = [doc.page_content for doc in docs]
vectors = embeddings.embed_documents(texts)  # list of lists

print(f'Documents embedded: {len(vectors)}')
print(f'Each vector dimensions: {len(vectors[0])}')
print(f'Total floats stored: {len(vectors) * len(vectors[0]):,}')
print(f'Storage per doc: ~{len(vectors[0]) * 4 / 1024:.1f} KB (float32)')
````
