# 🛡️ Guardrails

## What are Guardrails?
**Guardrails** = safety checks that prevent LLMs from producing harmful output.

## Types of Guardrails
- **Input guardrails** — filter harmful user input
- **Output guardrails** — filter harmful LLM output
- **Topic guardrails** — restrict LLM to specific topics
- **PII guardrails** — remove sensitive data
- **Hallucination detection** — fact-check LLM answers

## Production Necessity
Without guardrails:
- LLM can be jailbroken
- LLM can reveal sensitive data
- LLM can go off-topic
- Legal and reputational risk


```python
# ── Simple Input Guardrail ─────────────────────────────────────────────
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from pydantic import BaseModel, Field

class SafetyCheck(BaseModel):
    is_safe: bool = Field(description='Is the input safe and appropriate?')
    reason: str = Field(description='Reason for safety determination')
    category: str = Field(description='Category: safe, harmful, off_topic, pii')

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)
safety_checker = llm.with_structured_output(SafetyCheck)

def check_input_safety(user_input: str) -> SafetyCheck:
    return safety_checker.invoke(
        f'Analyze this user input for safety: "{user_input}"\n'
        'Determine if it is safe, harmful, off_topic, or contains PII.'
    )

# Test various inputs
test_inputs = [
    'Tell me about Python programming',
    'How do I hack into a bank?',
    'My SSN is 123-45-6789, help me',
]

for input_text in test_inputs:
    check = check_input_safety(input_text)
    status = '✅' if check.is_safe else '❌'
    print(f'{status} [{check.category}]: {input_text[:40]:40} | {check.reason[:60]}')
```
