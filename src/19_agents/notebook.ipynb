# 🤖 Agents

## What is an Agent?
An **agent** is an LLM that:
1. **Thinks** (decides what to do)
2. **Acts** (calls tools)
3. **Observes** (sees tool results)
4. **Repeats** until task complete

## ReAct Pattern (Reason + Act)
```
Thought: I need to find the weather in Paris
Action: get_weather(city="Paris")
Observation: Sunny, 22°C
Thought: I have the weather. I can answer now.
Final Answer: It's sunny and 22°C in Paris.
```

## Agent vs Chain
| Chain | Agent |
|-------|-------|
| Fixed steps | Dynamic steps |
| Predictable | Adaptive |
| Fast | Slower (multiple LLM calls) |
| Simple tasks | Complex reasoning |

## Modern Agents in LangChain
The modern way to build agents is with **LangGraph**.
But understanding legacy agents helps understand the concepts.


````python
# ── Building an Agent with create_react_agent ─────────────────────────
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langgraph.prebuilt import create_react_agent

@tool
def search_web(query: str) -> str:
    """Search the internet for current information on any topic."""
    # Simulated — in production use Tavily, DuckDuckGo, etc.
    fake_results = {
        'python': 'Python 3.12 released Oct 2023. New features: improved error messages.',
        'langchain': 'LangChain v0.3 released 2024. Major LCEL improvements.',
        'openai': 'GPT-4o released May 2024. Multimodal, faster, cheaper.',
    }
    for key, val in fake_results.items():
        if key in query.lower():
            return val
    return f'Search results for: {query} — No specific results found.'

@tool
def calculate(expression: str) -> str:
    """Evaluate mathematical expressions. Input: valid Python math expression."""
    try:
        return str(eval(expression, {'__builtins__': {}}))
    except:
        return 'Invalid expression'

tools = [search_web, calculate]
llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)

# create_react_agent: LangGraph-powered agent
agent = create_react_agent(llm, tools)

# Run the agent
result = agent.invoke({
    'messages': [('user', 'Search for info about Python 3.12 and tell me what 15 * 23 is.')]
})

# Extract final answer
print('Agent conversation:')
for msg in result['messages']:
    print(f'  [{msg.type}]: {str(msg.content)[:100]}')
````
