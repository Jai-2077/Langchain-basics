# 📄 Document Loaders

## What are Document Loaders?
Document Loaders **read data** from various sources and convert it into LangChain `Document` objects.

## The Document Object
```python
Document(
    page_content="The actual text content...",
    metadata={"source": "file.pdf", "page": 1}
)
```

## Why Document Loaders?
Your AI app needs to process:
- PDFs, Word docs, PowerPoints
- Websites, Notion, Confluence
- CSV, JSON, Excel files
- GitHub repos, YouTube transcripts
- SQL databases

Each source has a different format. Loaders normalize everything into `Document` objects.

## Available Loaders (100+)
| Loader | Source |
|--------|--------|
| PyPDFLoader | PDF files |
| WebBaseLoader | Websites |
| CSVLoader | CSV files |
| TextLoader | Plain text |
| DirectoryLoader | Folders |
| GitLoader | GitHub repos |
| YoutubeLoader | YouTube transcripts |


````python
# ── The Document Object ─────────────────────────────────────────────────
# WHY: All loaders produce Documents. Understanding Document = understanding RAG.

from langchain_core.documents import Document

# Create a Document manually
doc = Document(
    page_content='LangChain is a framework for building LLM applications.',
    metadata={
        'source': 'documentation.pdf',
        'page': 1,
        'author': 'Harrison Chase',
        'date': '2024-01-01'
    }
)

print('Type:', type(doc).__name__)
print('Content:', doc.page_content)
print('Metadata:', doc.metadata)
print('\nMetadata is passed all the way to the final answer!')
print('It lets you cite sources in RAG applications.')
````

````python
# ── TextLoader ─────────────────────────────────────────────────────────
# Simplest loader — loads plain text files.

import os
from langchain_community.document_loaders import TextLoader

# Create a sample file
sample_text = '''LangChain Documentation\n\nLangChain is a framework that enables developers to build applications\npowered by large language models.\n\nKey features:\n- LCEL for composing chains\n- Built-in memory\n- 100+ document loaders\n- Vector store integrations\n'''

with open('/tmp/sample.txt', 'w') as f:
    f.write(sample_text)

# Load it
loader = TextLoader('/tmp/sample.txt')
docs = loader.load()  # Returns list[Document]

print(f'Number of documents: {len(docs)}')
print(f'Content preview: {docs[0].page_content[:100]}')
print(f'Metadata: {docs[0].metadata}')
````

````python
# ── WebBaseLoader ───────────────────────────────────────────────────────
# Loads content from web URLs.
# WHY: Chat with any website! Foundation of website chatbots.

from langchain_community.document_loaders import WebBaseLoader

# Load a webpage
loader = WebBaseLoader(
    web_paths=['https://python.org'],
    # bs_kwargs={'parse_only': bs4.SoupStrainer('main')}  # filter HTML
)
docs = loader.load()

print(f'Loaded {len(docs)} document(s)')
print(f'Source: {docs[0].metadata["source"]}')
print(f'Content preview (first 200 chars):')
print(docs[0].page_content[:200])
````

````python
# ── CSVLoader ─────────────────────────────────────────────────────────
# Loads CSV files — each ROW becomes a Document.

import csv
from langchain_community.document_loaders import CSVLoader

# Create sample CSV
sample_csv = 'name,role,company\nAlice,Engineer,Google\nBob,Manager,Amazon\nCarol,Designer,Apple\n'
with open('/tmp/sample.csv', 'w') as f:
    f.write(sample_csv)

loader = CSVLoader('/tmp/sample.csv')
docs = loader.load()

print(f'Rows loaded as Documents: {len(docs)}')
for i, doc in enumerate(docs):
    print(f'\nDocument {i+1}:')
    print(f'  Content: {doc.page_content}')
    print(f'  Metadata: {doc.metadata}')
````
