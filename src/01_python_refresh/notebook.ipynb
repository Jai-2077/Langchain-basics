```python
# ── Type Hints ────────────────────────────────────────────────────────────
# WHY: LangChain is fully typed. Understanding types = understanding APIs.

from typing import Any, Dict, List, Optional, Union, Tuple

# Simple function with type hints
def greet(name: str, times: int = 1) -> str:
    """Returns greeting string.
    
    Args:
        name: Person's name
        times: How many times to repeat
    Returns:
        Greeting string
    """
    return (f'Hello, {name}! ' * times).strip()

result: str = greet('AI Engineer', 3)
print(result)
print(type(result))  # Always inspect types!
```

```python
# ── Pydantic Models ────────────────────────────────────────────────────────
# WHY: LangChain's ENTIRE data model is built on Pydantic.
# Every LangChain object (Messages, Documents, etc.) IS a Pydantic model.
# Understanding Pydantic = understanding LangChain internals.

from pydantic import BaseModel, Field, field_validator

class ChatMessage(BaseModel):
    """A simplified version of what LangChain's HumanMessage looks like internally."""
    role: str = Field(description='Who sent the message: user, assistant, system')
    content: str = Field(description='The actual message text')
    timestamp: Optional[str] = Field(default=None, description='When sent')

    @field_validator('role')
    @classmethod
    def role_must_be_valid(cls, v: str) -> str:
        allowed = {'user', 'assistant', 'system'}
        if v not in allowed:
            raise ValueError(f'role must be one of {allowed}')
        return v

# Create instance
msg = ChatMessage(role='user', content='Hello, AI!')
print(msg)               # Pretty print
print(msg.model_dump())  # → dict (used everywhere in LangChain)
print(msg.model_dump_json(indent=2))  # → JSON string
```

```python
# ── Generators & Streaming ─────────────────────────────────────────────────
# WHY: LangChain streaming uses Python generators under the hood.
# When you call chain.stream(), you get a generator.
# Understanding generators = understanding streaming.

from typing import Generator, Iterator

def fake_llm_stream(prompt: str) -> Generator[str, None, None]:
    """Simulates how LangChain streams tokens from an LLM."""
    words = f'Response to: {prompt}'.split()
    for word in words:
        yield word + ' '  # Each yield = one token chunk

# Consuming the generator (exactly how LangChain streaming works)
stream = fake_llm_stream('What is AI?')
print(f'Type: {type(stream)}')  # Generator object

full_response = ''
for chunk in stream:  # This is what chain.stream() does
    print(chunk, end='', flush=True)
    full_response += chunk
print(f'\n\nFull: {full_response}')
```

```python
# ── Async/Await ─────────────────────────────────────────────────────────────
# WHY: Production LangChain apps use async for performance.
# Every LangChain Runnable has .ainvoke(), .astream(), .abatch()
# 'a' prefix = async version.

import asyncio
from typing import AsyncGenerator

async def fake_async_llm(prompt: str) -> str:
    """Simulates async LLM call. In real code: await llm.ainvoke(prompt)"""
    await asyncio.sleep(0.01)  # Simulate network latency
    return f'[Async Response to: {prompt}]'

async def process_multiple_prompts(prompts: list[str]) -> list[str]:
    """Run multiple LLM calls concurrently. This is what .abatch() does."""
    # asyncio.gather runs all coroutines CONCURRENTLY (not sequentially)
    responses = await asyncio.gather(*[fake_async_llm(p) for p in prompts])
    return list(responses)

# Run async code
prompts = ['Tell me about Python', 'Tell me about AI', 'Tell me about LangChain']
results = asyncio.run(process_multiple_prompts(prompts))
for r in results:
    print(r)
```

```python
# ── Environment Variables ─────────────────────────────────────────────────
# WHY: API keys MUST never be hardcoded. LangChain reads keys from env vars.
# OPENAI_API_KEY, ANTHROPIC_API_KEY, LANGCHAIN_API_KEY etc.

import os
from dotenv import load_dotenv  # pip install python-dotenv

# load_dotenv() reads .env file and sets environment variables
load_dotenv()  # Always call this at the top of your scripts!

# Access env vars safely
api_key = os.getenv('OPENAI_API_KEY', 'not-set')
print(f'Key loaded: {api_key[:8]}...' if api_key != 'not-set' else 'Key not set - check .env')

# Pattern used throughout this course
def get_required_env(key: str) -> str:
    value = os.getenv(key)
    if not value:
        raise EnvironmentError(f'Missing required env var: {key}\nAdd it to your .env file')
    return value
```
