# 🔗 LangChain Basics

## What is LangChain?
LangChain is a **framework** (collection of tools) that makes building LLM-powered applications easier.

## The Problem LangChain Solves
Without LangChain, building an AI app looks like this:
```
1. Manually format prompts as strings
2. Call OpenAI API directly
3. Parse the JSON response manually
4. No retry logic, no streaming, no memory
5. Repeat for every feature you need
```

With LangChain:
```
chain = prompt | llm | parser
result = chain.invoke({"input": "Hello"})
```

## LangChain's 3 Core Packages
| Package | Purpose |
|---------|---------|
| `langchain-core` | Base interfaces (Runnable, BaseMessage, etc.) |
| `langchain` | High-level chains, agents, retrievers |
| `langchain-openai` | OpenAI integration |

## The LCEL Philosophy
**L**ang**C**hain **E**xpression **L**anguage = pipe operator `|`
```
component_a | component_b | component_c
```
Inspired by Linux pipes: `cat file.txt | grep "error" | wc -l`


````python
# ── LangChain Package Structure ───────────────────────────────────────────
# WHY so many packages? Modular design = install only what you need.

import langchain
import langchain_core
import langchain_openai

print('LangChain version:', langchain.__version__)
print('LangChain Core version:', langchain_core.__version__)

# Key modules you'll use constantly
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI

print('\nAll imports successful!')
print('These 5 imports cover 80% of LangChain usage.')
````

````python
# ── The Simplest Possible Chain ───────────────────────────────────────────
# BEFORE LangChain:
import openai
import os

# Old way: manual API calls
def old_way(topic: str) -> str:
    client = openai.OpenAI()
    response = client.chat.completions.create(
        model='gpt-4o-mini',
        messages=[{'role': 'user', 'content': f'Tell me about {topic}'}]
    )
    return response.choices[0].message.content

# ─────────────────────────────────────────────────────────────────────────
# AFTER LangChain: cleaner, composable, swappable
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

def langchain_way(topic: str) -> str:
    prompt = ChatPromptTemplate.from_template('Tell me about {topic}')
    llm = ChatOpenAI(model='gpt-4o-mini')
    parser = StrOutputParser()

    # The pipe | operator chains these together
    chain = prompt | llm | parser
    return chain.invoke({'topic': topic})

print('LangChain way is:')
print('  ✓ More readable')
print('  ✓ Composable (swap any component)')
print('  ✓ Streaming built-in')
print('  ✓ Async built-in')
print('  ✓ Tracing built-in')
````

````python
# ── Anatomy of a Chain ─────────────────────────────────────────────────────
# Let's trace EXACTLY what happens inside chain.invoke()

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

topic = 'Python programming'

# Step 1: PromptTemplate formats your variables into a structured prompt
prompt = ChatPromptTemplate.from_template('Explain {topic} in one sentence.')
step1_output = prompt.invoke({'topic': topic})
print('After PromptTemplate:')
print(f'  Type: {type(step1_output).__name__}')  # ChatPromptValue
print(f'  Messages: {step1_output.messages}')

# Step 2: LLM takes messages, returns AIMessage
llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)
step2_output = llm.invoke(step1_output)
print(f'\nAfter LLM:')
print(f'  Type: {type(step2_output).__name__}')  # AIMessage
print(f'  Content: {step2_output.content[:80]}...')

# Step 3: StrOutputParser extracts .content → plain string
parser = StrOutputParser()
step3_output = parser.invoke(step2_output)
print(f'\nAfter Parser:')
print(f'  Type: {type(step3_output).__name__}')  # str
print(f'  Value: {step3_output[:80]}...')
````
