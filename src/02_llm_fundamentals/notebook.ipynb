# 🧠 LLM Fundamentals

## What is an LLM?
A **Large Language Model (LLM)** is a neural network trained on massive amounts of text to predict the next token.

## Mental Model: The Autocomplete Genius
Think of an LLM as the world's most sophisticated autocomplete:
- Your phone's autocomplete suggests the next word based on recent context
- An LLM suggests the next **token** based on ALL the text ever written

## Key Concepts
| Term | Meaning |
|------|---------|
| Token | ~4 characters or ~0.75 words |
| Context Window | Max tokens LLM can "see" at once |
| Temperature | Randomness (0=deterministic, 2=creative chaos) |
| Top-p | Nucleus sampling — limits token pool |
| System Prompt | Instructions that shape LLM behavior |
| Completion | LLM's response |

## Why LangChain Wraps LLMs
Raw LLM APIs are painful:
- Different providers = different APIs
- No memory between calls
- No easy chaining
- No streaming abstraction
- No retry logic

LangChain gives you **ONE interface** for ALL providers.


```python
# ── Understanding Tokens ──────────────────────────────────────────────────
# WHY: Token count determines cost and context limits.
# Every LLM has a max context window measured in TOKENS, not words.

# Install: pip install tiktoken
try:
    import tiktoken
    enc = tiktoken.encoding_for_model('gpt-4o')
    text = 'LangChain makes building LLM applications much easier!'
    tokens = enc.encode(text)
    print(f'Text: {text}')
    print(f'Token count: {len(tokens)}')
    print(f'Tokens: {tokens}')
    print(f'Token strings: {[enc.decode([t]) for t in tokens]}')
except ImportError:
    print('tiktoken not installed. Run: pip install tiktoken')
    print('Rule of thumb: 1 token ≈ 4 chars ≈ 0.75 words')
```

```python
# ── Your First LLM Call with LangChain ─────────────────────────────────────
# This is the SIMPLEST possible LangChain usage.
# We'll build on this pattern for the entire course.

import os
from dotenv import load_dotenv
load_dotenv()

from langchain_openai import ChatOpenAI  # pip install langchain-openai

# ── Step 1: Create LLM instance ─────────────────────────────────────────
# WHY ChatOpenAI not OpenAI?
# 'Chat' = uses chat completion API (messages format)
# Old 'OpenAI' class used text completion API (deprecated by OpenAI)
llm = ChatOpenAI(
    model='gpt-4o-mini',    # Model name
    temperature=0.7,         # Creativity: 0=deterministic, 1=creative
    max_tokens=500,          # Max response length
)

# ── Step 2: Call the LLM ─────────────────────────────────────────────────
# WHY .invoke()? Because all LangChain Runnables use .invoke().
# Consistent interface across all components.
response = llm.invoke('What is LangChain in one sentence?')

# ── Step 3: Inspect the response ─────────────────────────────────────────
print('Type:', type(response))    # AIMessage — NOT a string!
print('Content:', response.content)
print('Metadata:', response.response_metadata)
print('Token usage:', response.usage_metadata)
```

```python
# ── Temperature Experiment ─────────────────────────────────────────────────
# WHY does temperature matter?
# temperature=0 → same answer every time (good for facts, code)
# temperature=1 → creative, varied (good for writing, brainstorming)

from langchain_openai import ChatOpenAI

prompt = 'Give me a creative name for an AI startup. Just the name, nothing else.'

print('── temperature=0 (deterministic) ──')
llm_cold = ChatOpenAI(model='gpt-4o-mini', temperature=0)
for i in range(3):
    r = llm_cold.invoke(prompt)
    print(f'  Run {i+1}: {r.content}')  # Same every time

print('\n── temperature=1 (creative) ──')
llm_hot = ChatOpenAI(model='gpt-4o-mini', temperature=1)
for i in range(3):
    r = llm_hot.invoke(prompt)
    print(f'  Run {i+1}: {r.content}')  # Different every time
```
